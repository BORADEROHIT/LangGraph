```python
from langchain_aws import ChatBedrockConverse, BedrockEmbeddings
from langchain_classic.schema import SystemMessage, HumanMessage
# LangGraph
from typing import TypedDict, List
from langgraph.graph import StateGraph, END
#  Define LangGraph StateGraph agent
from langgraph.graph import StateGraph
# Create a LangGraph agent with memory and tools
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.agents import initialize_agent
from langchain_core.messages import HumanMessage
```

```python
#Configure AWS Bedrock credentials
# ---------------------------
# AWS Bedrock LLM Setup
# ---------------------------
llm = ChatBedrockConverse(model_id="amazon.nova-lite-v1:0", region_name="us-east-1", temperature=0.5, max_tokens=100)
```

```python
# Define state type
class AgentState(TypedDict):
    messages:List[HumanMessage]
# Define a simple node function
def respond(state: AgentState) -> AgentState:
    messages = state.get("messages", [])
    response = llm.invoke(messages)
    if response is None:
        raise ValueError("LLM returned None, check LLM initialization/API key.")
    messages.append(response)
    return AgentState({"messages": messages})
# Create graph
graph = StateGraph(AgentState)
graph.add_node("respond", respond)
graph.set_entry_point("respond")
workflow = graph.compile()
```

```python
#  Run the agent workflow
state = AgentState({"messages": [HumanMessage(content="Hello, what can you do?")]})
result = workflow.invoke(state)
print(result["messages"][-1].content)
```
