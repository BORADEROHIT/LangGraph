```python
from typing import TypedDict, List, Dict
import json
from langgraph.graph import StateGraph, END
from langchain_aws import ChatBedrockConverse
from langchain_core.prompts import ChatPromptTemplate
# ---------- LLM (Amazon Nova Lite) ----------
llm = ChatBedrockConverse(model_id="amazon.nova-lite-v1:0",
                          region_name="us-east-1",
                          temperature=0.2,
                          max_tokens=400,
                         )
```

```python
# ---------- State ----------
class State(TypedDict):
    objective: str
    constraints: str
    tasks:List [ Dict ]
    i: int
    executions:Dict[str, str]
```

```python
# ---------- Prompts (NOTE: double braces to escape JSON) ----------
planner_prompt = ChatPromptTemplate.from_messages([
    ("system","You are a planning assistant. Break the objective into 5-8 tasks.Each task MUST include:\
     task, priority (High|Medium|Low), and depends_on (array of task names).Return STRICT JSON only with this schema:"
     ),("human", "Objective: {objective}\nConstraints: {constraints}")])
executor_prompt = ChatPromptTemplate.from_messages([
    ("system","You are an execution assistant. For the given task, produce 3-5 concise, actionable steps.\
     Avoid dates or weekdays. Be specific and sequential"), ("human","Task: {task}\nPriority: {priority}\
     Depends on: {depends}Output format- Step 1: ...Step 2: ..." )
     ])
```

```python
# ---------- Nodes ----------
def plan_node(state: State) -> State:
    content={"objective": state["objective"], "constraints": state["constraints"]}
    msg = (planner_prompt | llm).invoke(content)
    text = getattr(msg, "content", "").strip()
    try:
        obj = json.loads(text)
        state["tasks"] = obj.get("tasks", [])
    except Exception:
        state["tasks"] = [{"task": "Define scope", "priority": "High", "depends_on": []},
                          {"task": "Collect requirements", "priority": "High", "depends_on": ["Define scope"]},
                          {"task": "Design solution outline", "priority": "Medium", "depends_on": ["Collect requirements"]},
                         ]
    state["i"] = 0
    state["executions"] = {}
    return state
```

```python
def execute_task_node(state: State) -> State:
    idx = state["i"]
    t = state["tasks"][idx]
    msg = (executor_prompt | llm).invoke({"task": t.get("task", f"Task {idx+1}"),"priority": t.get("priority", "Medium"),"depends": ", ".join(t.get("depends_on", [])) or "None" })
    content = getattr(msg, "content", "").strip()
    state["executions"][t.get("task", f"Task {idx+1}")] = content
    state["i"] = idx + 1
    return state
def route_next(state: State):
    return END if state["i"] >= len(state.get("tasks", [])) else "execute_task"
```

```python
# ---------- Build graph ----------
graph = StateGraph(State)
graph.add_node("plan", plan_node)
graph.add_node("execute_task", execute_task_node)
graph.set_entry_point("plan")
graph.add_conditional_edges("plan", lambda s: "execute_task", {"execute_task": "execute_task"})
graph.add_conditional_edges("execute_task", route_next, {"execute_task": "execute_task", END: END})
app = graph.compile()
```
